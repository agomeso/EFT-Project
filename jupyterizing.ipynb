```python
import pandas as pd

file = pd.read_csv('bestsellers.csv')

file
```

```python
file['Genre'].unique()
```

```python
file['Year'].sort_values().unique()
```

```python
file['Name'].unique()
```

```python
df = file.groupby(['Name'])
unique = df.max()
unique = unique.reset_index()
book_price = unique[['Name','Author','Price']]
book_price
```

```python
table = file[['Name','User Rating','Reviews','Year']]
table
```

```python
connection = unique[['Name','Genre']]
connection
```

```python
from sqlalchemy import create_engine
rds_connection_string = "postgres:postgres@localhost:5432/best_book_db"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
engine.table_names()
```

```python
book_price.to_sql(name='Price', con=engine, if_exists='append', index=False)
```

```python
table.to_sql(name='Reviews_and_Ratings', con=engine, if_exists='append', index=False)
```

```python
connection.to_sql(name='Genre', con=engine, if_exists='append', index=False)
```

```python
Price_df = pd.read_sql_query('select * from "Price"', con=engine)
```

```python
Reviews_and_Ratings_df = pd.read_sql_query('select * from "Reviews_and_Ratings"', con=engine)
```

```python
Genre_df = pd.read_sql_query('select * from "Genre"', con=engine)
```

```python
new_df = pd.merge(Reviews_and_Ratings_df,Price_df,on="Name",how="left")
```

```python
combined_df = pd.merge(new_df,Genre_df,on="Name",how="left")
combined_df
```
